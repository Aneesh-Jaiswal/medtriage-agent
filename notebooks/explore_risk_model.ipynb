# Explore Risk Classifier

Quick notebook for loading synthetic patient data, inspecting the feature
distribution, and evaluating the trained `RiskClassifier` checkpoint.

Run `python data/synthetic/generate_patients.py` and `python models/train.py`
before using this notebook.

```python
import sys
sys.path.append('..')

import pandas as pd
import torch

from models.features import featurize, feature_dim
from models.risk_classifier import load_model

df = pd.read_csv('../data/synthetic/patients.csv')
df.head()
```

```python
df['risk_label'].value_counts()
```

```python
model = load_model('../models/checkpoints/risk_classifier.pt', input_dim=feature_dim())

sample = df.iloc[0]
x = featurize(int(sample['age']), int(sample['duration_hours']), sample['symptom_text'])
print('Predicted:', model.predict_label(x.unsqueeze(0)))
print('Actual:', sample['risk_label'])
```
